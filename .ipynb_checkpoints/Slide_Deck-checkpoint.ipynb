# Do all students compete on an equal footing ?


### THE ANALYSIS

I will try to answer this question by analyzing factors that are outside of the student's control, and visualizing their impact on the grades of the students.<br>

Perhaps the most obvious one is the __impact of the parent'socio-economic characteristics__ on the  student's grades. The second aspect,which was popularized by Malcom Gladwell in his book outliers, is the  __relationship between the student's age and his grades__. Finally, we are going to visualize the impact of __having the mother and/or the father at home__.<br>

The analysis will be done with data coming from the __2012 PISA dataset__.

### THE PISA DATASET

PISA is a survey of students' skills and knowledge as they approach the end of compulsory education. It is not a conventional school test. Rather than examining how well students have learned the school curriculum, it looks at how well prepared they are for life beyond school.

Around 510,000 students in 65 economies took part in the PISA 2012 assessment of reading, mathematics and science representing about 28 million 15-year-olds globally. Of those economies, 44 took part in an assessment of creative problem solving and 18 in an assessment of financial literacy.
<br>

```python
# Importing the relevant librairies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sb
```

```python
#loading the dataset
pisa = pd.read_csv('pisa2012.csv', encoding = 'cp1252', low_memory = False, index_col = 0)
```

# The impact of the parent's socio-economic characteristics on the  student's grades. 

## Higher Socio-economic status -- higher grades

The average SQ ISEI whether the mother's, the father's or combined (a quantitative proxy for the socio economic status) is correlated grades. <br>
Perhaps even more interestingly, is that the combined SQ ISEI has a smaller interquartile range, or in other words is more tightly correlated with the grades.



### What is SQ ISEI ?

The Programme for International Student Assessment (PISA) International Socio-Economic Index of Occupational Status (ISEI) was derived from students’ responses on parental occupation. The index captures the attributes of occupations that convert parents’ education into income. The index was derived by the optimal scaling of occupation groups to maximise the indirect effect of education on income through occupation and to minimise the direct effect of education on income, net of occupation (both effects being net of age).

```python
pisa['SQISEI_comb'] = pisa['BFMJ2'] + pisa['BMMJ1']
fig = plt.figure(figsize = (20,10))

ax1 = fig.add_subplot(231)
sb.barplot(data = pisa,x = 'GRADE', y='BFMJ2', ci = 'sd', color = sb.color_palette()[0])
plt.title('Father\'s Socio-Economic Characteristics as a function of the grades',loc ='center')
plt.xlabel('Grade compared to the modal grade in country')
plt.ylabel('SQ ISEI: Father\'s Socio-Economic Characteristics')
ttl = ax1.title
ttl.set_position([.45, 1.05])

ax2 = fig.add_subplot(232)
sb.barplot(data = pisa,x = 'GRADE', y='BMMJ1', color = sb.color_palette()[0], ci = 'sd')
plt.title('Mother\'s Socio-Economic Characteristics as a function of the grades',loc ='center')
plt.xlabel('Grade compared to the modal grade in country')
plt.ylabel('SQ ISEI: Mother\'s Socio-Economic Characteristics')
ttl = ax2.title
ttl.set_position([.45, 1.05])

ax3 = fig.add_subplot(233)
sb.barplot(data = pisa,x = 'GRADE', y='SQISEI_comb', color = sb.color_palette()[0], ci = 'sd')
plt.title('Combined (father/mother) Socio-Economic Characteristics as a function of the grades',loc ='center')
plt.xlabel('Grade compared to the modal grade in country')
plt.ylabel('SQ ISEI: Combined Socio-Economic Characteristics.')
ax = plt.gca()
ttl = ax.title
ttl.set_position([.55, 1.05])

ax4 = fig.add_subplot(234)
sb.boxplot(data = pisa,x = 'GRADE', y='BFMJ2', color = sb.color_palette()[0])
plt.xlabel('Grade compared to the modal grade in country')
plt.ylabel('SQ ISEI: Father\'s Socio-Economic Characteristics')

ax5 = fig.add_subplot(235)
sb.boxplot(data = pisa,x = 'GRADE', y='BMMJ1', color = sb.color_palette()[0])
plt.xlabel('Grade compared to the modal grade in country')
plt.ylabel('SQ ISEI: Father\'s Socio-Economic Characteristics')

ax6 = fig.add_subplot(236)
sb.boxplot(data = pisa,x = 'GRADE', y='SQISEI_comb', color = sb.color_palette()[0])
plt.xlabel('Grade compared to the modal grade in country')
plt.ylabel('SQ ISEI: Father\'s Socio-Economic Characteristics')

plt.suptitle('Are socio-economic characteristics linked to the student\'s grade ?', x = 0.5, y = 1.05, fontsize = 20, fontweight='bold', color = 'blue')
plt.tight_layout()
plt.show()
```

# The impact of having the mother and/or the father at home.

### Engineered features

I created two variables sense_of_belonging (which I will use as a proxy for social affinity) and perceived_control (which I will use as a proxy for internal locus of control). This was meant to simply to be able to more easily compare these metrics with the presence of the father or mother at home.

## Mother/Father where are you ?

Students with either mother or father at home have an increased sense of belonging and perceived control. Social affinity and high internal locus are usually described as traits of  of highly succesful individuals.


```python
# Create a variable for overall perceived control

#map the categorical values to their corresponding value
agreement_dict = {'Strongly disagree':-2,'Disagree':-1,'Agree':1,'Strongly agree':2}
agreement_dict_r = {'Strongly disagree':2,'Disagree':1,'Agree':-1,'Strongly agree':-2}

#Creat a positive and negative dictionnary for the perceived control markers
positive_columnlist = [['pc_effort','ST43Q01'],
                       ['pc_doingwell','ST43Q02'],
                       ['pc_willtoperform','ST43Q05']]


negative_columnlist =[['pc_family_demands','ST43Q04'],
                      ['pc_different_teachers','ST43Q05'],
                      ['perform_poorly','ST43Q06']]

#map each of the values in a column
for newcolumn,mappedcolumn in positive_columnlist:
    pisa[newcolumn] = pisa[mappedcolumn].map(agreement_dict)

for newcolumn,mappedcolumn in negative_columnlist:
    pisa[newcolumn] = pisa[mappedcolumn].map(agreement_dict_r)

#Create the perceived control variable
pisa['perceived_control'] = pisa['pc_effort'] + pisa['pc_doingwell'] + pisa['pc_willtoperform'] + pisa['pc_family_demands'] + pisa['pc_different_teachers'] + pisa['perform_poorly']
```

```python
#Create a new variable sense of belonging
agreement_dict = {'Strongly disagree':-2,'Disagree':-1,'Agree':1,'Strongly agree':2}

columnlist = [['feel_outsider','ST87Q01'],
              ['makes_friends','ST43Q02'],
              ['belong_at_school','ST87Q03'],
              ['awkward_at_school','ST87Q04'],
              ['liked_by_others','ST87Q05'],
              ['feel_lonely','ST87Q06'],
              ['feel_happy','ST87Q07'],
              ['ideal_at_school','ST87Q08'],
              ['satisfied_at_school','ST87Q09']]
                       
for newcolumn,mappedcolumn in columnlist:
    pisa[newcolumn] = pisa[mappedcolumn].map(agreement_dict)

pisa['sense_of_belonging'] = - pisa['feel_outsider'] + pisa['makes_friends'] + pisa['belong_at_school'] - pisa['awkward_at_school'] + pisa['liked_by_others'] - pisa['feel_lonely'] + pisa['feel_happy'] + pisa['ideal_at_school'] + pisa['satisfied_at_school']
```

```python
plt.figure(figsize = (10,8))
pisa_sub = pisa.groupby(['ST11Q02'],sort = False)
pisa_count = pisa_sub.perceived_control.value_counts(normalize = True, sort = False)
pisa_sub_count = [{'ST11Q02': ST11Q02, 'perceived_control':perceived_control, 'percentage': percentage*100} for (ST11Q02,perceived_control), percentage in dict(pisa_count).items()]
pisa_sub_df = pd.DataFrame(pisa_sub_count) 

ax = sb.barplot(data = pisa_sub_df, x ='perceived_control', y = 'percentage', hue = 'ST11Q02')
plt.legend(title = 'Father at home')
plt.title('Sense of perceived control proportions with/without Father')
plt.ylabel('Percentage')
plt.xlabel('Perceived Control')
step = 2
ax.set(xticks = ax.get_xticks()[::step], xticklabels= ax.get_xticklabels()[::step])
plt.show()
```

```python
plt.figure(figsize = (10,8))
pisa_sub = pisa.groupby(['ST11Q02'],sort = False)
pisa_count = pisa_sub.sense_of_belonging.value_counts(normalize = True, sort = False)
pisa_sub_count = [{'ST11Q02': ST11Q02, 'sense_of_belonging':sense_of_belonging, 'percentage': percentage*100} for (ST11Q02,sense_of_belonging), percentage in dict(pisa_count).items()]
pisa_sub_df = pd.DataFrame(pisa_sub_count) 

ax = sb.barplot(data = pisa_sub_df, x ='sense_of_belonging', y = 'percentage', hue = 'ST11Q02')
plt.legend(title = 'Father at home')
plt.title('Sense of belonging proportions with/without Father')
plt.ylabel('Percentage')
plt.xlabel('Sense of belonging')
step = 6
ax.set(xticks = ax.get_xticks()[::step], xticklabels= ax.get_xticklabels()[::step])
plt.show()
```

```python
plt.figure(figsize = (10,8))
pisa_sub = pisa.groupby(['ST11Q01'],sort = False)
pisa_count = pisa_sub.sense_of_belonging.value_counts(normalize = True, sort = False)
pisa_sub_count = [{'ST11Q01': ST11Q01, 'sense_of_belonging':sense_of_belonging, 'percentage': percentage*100} for (ST11Q01,sense_of_belonging), percentage in dict(pisa_count).items()]
pisa_sub_df = pd.DataFrame(pisa_sub_count) 

ax = sb.barplot(data = pisa_sub_df, x ='sense_of_belonging', y = 'percentage', hue = 'ST11Q01')
plt.legend(title = 'Mother at home')
plt.title('Sense of belonging proportions with/without Mother')
step = 6
ax.set(xticks = ax.get_xticks()[::step], xticklabels= ax.get_xticklabels()[::step])
plt.show()
```

## Mother and Father always make me do my homework! 

A higher proportion of students with either their mother or father at home have higher grades. In general, the presence of the father or the mother at home have a positive impact on the grades of the students.

```python
# dataframe (it is used to count the number of students in the groups)
df_new = pisa[["GRADE", "ST11Q01"]]
df_new.reset_index(inplace = True)
df_new.columns=['ind','Grade','Mother\nAt Home']
# count the number of students in each group
df_new_per = df_new.groupby(["Mother\nAt Home","Grade"]).agg({'ind': 'count'})

# groupby ST11Q01 (the first level in the index) and divide by sum
grade_pcts = df_new_per.groupby(level=0).apply(lambda x:
                                                 100 * x / float(x.sum())).reset_index()

# plot
plt.figure(figsize=(7.5,7.5))
sb.barplot(x='Grade', y='ind' , hue = "Mother\nAt Home", data=grade_pcts);
plt.legend(ncol = 2, title = 'Mother at home' )
plt.title('Percentage Grade distribution with/without mother at home')
plt.show()
```

```python
# dataframe (it is used to count the number of students in the groups)
df_new = pisa[["GRADE", "ST11Q02"]]
df_new.reset_index(inplace = True)
df_new.columns=['ind','Grade','Father\nAt Home']
# count the number of students in each group
df_new_per = df_new.groupby(["Father\nAt Home","Grade"]).agg({'ind': 'count'})

# groupby ST11Q02 (the first level in the index) and divide by sum
grade_pcts = df_new_per.groupby(level=0).apply(lambda x:
                                                 100 * x / float(x.sum())).reset_index()

# plot
plt.figure(figsize=(7.5,7.5))
sb.barplot(x='Grade', y='ind' , hue = "Father\nAt Home", data=grade_pcts);
plt.legend(ncol = 2, title = 'Father at home' )
plt.title('Percentage Grade distribution with/without father at home')
plt.show()
```

## Mother vs. Father

Students generally have better grades when both parents are at home rather than none. <br>

More importantly this graph shows that having the mother at home has a bigger positive impact on the grades than having the father at home.


```python
# dataframe (it is used to count the number of students in the groups)
df_new = pisa[["GRADE", "ST11Q01",'ST11Q02']]
df_new.reset_index(inplace = True)
df_new.columns=['ind','Grade','Mother\nAt Home','Father\nAt Home']

# count the number of students in each group
parents_at_home = df_new.groupby(["Mother\nAt Home",'Father\nAt Home']).mean()

parents_at_home.reset_index(inplace = True)

piv_parents_at_home = pd.pivot_table(parents_at_home,
                                    values = 'Grade',
                                    index = ['Mother\nAt Home'],
                                    columns = ['Father\nAt Home'],
                                    fill_value = 0)


plt.figure(figsize = (7.5,7.5))
sb.heatmap(piv_parents_at_home, annot = True)
plt.title('Heat map depicting the average grade based on whether the father/mother are present at home.')
ax = plt.gca()
ttl = ax.title
ttl.set_position([.5, 1.05])
plt.show()
```

# The relationship between the students age and their grades.

## "Intelligence is in the head, not in the age" ... Or is it ? 

The most interesting aspect of my analysis was that the age of the student is highly correlated with his grades.

```python
plt.figure(figsize=(7.5,7.5))
sb.barplot(data = pisa,x = 'AGE', y='GRADE', color = sb.color_palette()[0])
plt.title('Average grade as a function of the age of a student',loc ='center')
plt.ylabel('Average grade compared to the modal grade in country')
plt.xlabel('Age')
ax = plt.gca()
ttl = ax.title
ttl.set_position([.5, 1.05])
plt.tight_layout()
plt.show()
```

## Even clearer...

If we go even further, and remove all the students that repeated a year, the correlation between age and grades is even clearer. 

```python
pisa_no_repeat = pisa.query('REPEAT == "Did not repeat a <grade>"')
plt.figure(figsize=(7.5,7.5))
sb.barplot(data = pisa_no_repeat,x = 'AGE', y='GRADE', color = sb.color_palette()[0])
plt.title('Average grade as a function of the age of a student',loc ='center')
plt.ylabel('Average grade compared to the modal grade in country')
plt.xlabel('Age')
ax = plt.gca()
ttl = ax.title
ttl.set_position([.5, 1.05])
plt.tight_layout()
plt.show()
```
