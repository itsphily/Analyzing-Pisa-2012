# THE PISA DATASET

PISA is a survey of students' skills and knowledge as they approach the end of compulsory education. It is not a conventional school test. Rather than examining how well students have learned the school curriculum, it looks at how well prepared they are for life beyond school.

Around 510,000 students in 65 economies took part in the PISA 2012 assessment of reading, mathematics and science representing about 28 million 15-year-olds globally. Of those economies, 44 took part in an assessment of creative problem solving and 18 in an assessment of financial literacy.

####  <font color = 'red'> Geography </font>
>- There is about 100,000 more students that have been surveyed in OECD countries compared to non-OECD countries. I will have to be careful when comparing the two.

#### <font color = 'red'> Grades </font>
>- Grade distribution compared to the modal grade in all the  countries is roughly normal. 

#### <font color = 'red'> Parents </font>
>- Both parents have almost the same level of education, with a majority of the parents that have completed ISCED level 3A, 3B, and 3C.

#### <font color = 'red'> Books </font>
>- The number of books owned has a right skewed distribution, with an average of 26-100 books per household.

#### <font color = 'red'> Perceived control </font>
>- In general students tend to have high internal locus of control. They feel they have control over the grades they could potentially attain. However, they still believe that with different teachers they could get better grades. It would be interesting to see if the quality of the teachers or the perception of the quality of the teacher has an impact on grades. It will also be interesting to see if the students with higher internal locus tend to have better grades. 

#### <font color = 'red'> Feelings towards math </font>
>- If we look at the overal metrics that qualify the relationship to Math, there isn't really a trend that we can distinguish. We would need to dig deeper if we wanted to make any statements with respect to this data.

```python
# dataframe (it is used to count the number of students in the groups)
df_new = pisa[["GRADE", "ST11Q01"]]
df_new.reset_index(inplace = True)
df_new.columns=['ind','Grade','Mother\nAt Home']
# count the number of students in each group
df_new_per = df_new.groupby(["Mother\nAt Home","Grade"]).agg({'ind': 'count'})

# groupby ST11Q01 (the first level in the index) and divide by sum
grade_pcts = df_new_per.groupby(level=0).apply(lambda x:
                                                 100 * x / float(x.sum())).reset_index()

# plot
sb.barplot(x='Grade', y='ind' , hue = "Mother\nAt Home", data=grade_pcts);
plt.legend(ncol = 2, title = 'Mother at home' )
plt.title('Percentage Grade distribution with/without mother at home')
plt.show()
```

```python
# dataframe (it is used to count the number of students in the groups)
df_new = pisa[["GRADE", "ST11Q02"]]
df_new.reset_index(inplace = True)
df_new.columns=['ind','Grade','Father\nAt Home']
# count the number of students in each group
df_new_per = df_new.groupby(["Father\nAt Home","Grade"]).agg({'ind': 'count'})

# groupby ST11Q02 (the first level in the index) and divide by sum
grade_pcts = df_new_per.groupby(level=0).apply(lambda x:
                                                 100 * x / float(x.sum())).reset_index()

# plot
sb.barplot(x='Grade', y='ind' , hue = "Father\nAt Home", data=grade_pcts);
plt.legend(ncol = 2, title = 'Father at home' )
plt.title('Percentage Grade distribution with/without father at home')
plt.show()
```

```python
agreement_dict = {'Strongly disagree':-2,'Disagree':-1,'Agree':1,'Strongly agree':2}
agreement_dict_r = {'Strongly disagree':2,'Disagree':1,'Agree':-1,'Strongly agree':-2}

#Creat a positive and negative dictionnary for the perceived control markers
positive_columnlist = [['pc_effort','ST43Q01'],
                       ['pc_doingwell','ST43Q02'],
                       ['pc_willtoperform','ST43Q05']]


negative_columnlist =[['pc_family_demands','ST43Q04'],
                      ['pc_different_teachers','ST43Q05'],
                      ['perform_poorly','ST43Q06']]

#map each of the values in a column
for newcolumn,mappedcolumn in positive_columnlist:
    pisa[newcolumn] = pisa[mappedcolumn].map(agreement_dict)

for newcolumn,mappedcolumn in negative_columnlist:
    pisa[newcolumn] = pisa[mappedcolumn].map(agreement_dict_r)
```

####  <font color = 'red'> Geography </font>
>- Male and Female students are more equally represented in OECD countries, whereas in Non-OECD countries there is a slightly  bigger difference in attendance. The difference is 2.44% in non-OECD countries versus 0.0003% in OECD countries .<br>
>- The number of books at home is significantly higher in OECD countries. 
> __Note: Most OECD members are high-income economies with a very high Human Development Index (HDI) and are regarded as developed countries.__

#### <font color = 'red'> Parents </font>
>- The presence of the father or the mother at home have a positive impact on the grades of the students. A higher proportion of students with either their mother or father at home have higher grades. It would be interesting to compare the effect of the mother and the father on the grades of the students

#### <font color = 'red'> Grades </font>
>- Grade distribution compared to the modal grade in all the countries is roughly normal. 
>- Both male and female students show approximately a normal distribution in grades. However female students tend to have higher grades (-.20 for males, -.126 for females).
>- We can see that the average SQ ISEI (which is a quantitative proxy for the socio economic characteristic) tends to increase with higher grades with respect to the average. Whether it is the father's, the mother's or the combined SQ ISEI. This trend is still present. Perhaps even more interestingly (even though it seems obvious), is that the combined SQ ISEI has a smaller interquartile range, or in other words is more tightly correlated with the grades. It would be interesting to see how far this correlation goes by cleaning our data for instance taking the combined SQ ISEI only when the parents are still together. 
>- Perhaps the most interesting aspect of my analysis was that the age of the student is highly correlated with his grades. If we go even further, and remove all the students that repeated a year, the correlation is even clearer. 

> __Note: The Programme for International Student Assessment (PISA) International Socio-Economic Index of Occupational Status (ISEI) was derived from students’ responses on parental occupation. The index captures the attributes of occupations that convert parents’ education into income. The index was derived by the optimal scaling of occupation groups to maximise the indirect effect of education on income through occupation and to minimise the direct effect of education on income, net of occupation (both effects being net of age).__ 

#### <font color = 'red'> Perceived Control and sense of belonging </font>

__I created two variables sense_of_belonging (which I will use as a proxy for social affinity) and perceived_control (which I will use as a proxy for internal locus of control). >- In general students with either mother or father at home have an increased sense_of_belonging and perceived_control. However there is something interesting in the perceived control distribution. Proving that my mom smothering me with love and protecting me from everything wasn’t necessarily in my benefit : )

### Feature engineering <a id='feature_engineering'></a>

```python
# Create a variable for overall perceived control

#map the categorical values to their corresponding value
agreement_dict = {'Strongly disagree':-2,'Disagree':-1,'Agree':1,'Strongly agree':2}
agreement_dict_r = {'Strongly disagree':2,'Disagree':1,'Agree':-1,'Strongly agree':-2}

#Creat a positive and negative dictionnary for the perceived control markers
positive_columnlist = [['pc_effort','ST43Q01'],
                       ['pc_doingwell','ST43Q02'],
                       ['pc_willtoperform','ST43Q05']]


negative_columnlist =[['pc_family_demands','ST43Q04'],
                      ['pc_different_teachers','ST43Q05'],
                      ['perform_poorly','ST43Q06']]

#map each of the values in a column
for newcolumn,mappedcolumn in positive_columnlist:
    pisa[newcolumn] = pisa[mappedcolumn].map(agreement_dict)

for newcolumn,mappedcolumn in negative_columnlist:
    pisa[newcolumn] = pisa[mappedcolumn].map(agreement_dict_r)

#Create the perceived control variable
pisa['perceived_control'] = pisa['pc_effort'] + pisa['pc_doingwell'] + pisa['pc_willtoperform'] + pisa['pc_family_demands'] + pisa['pc_different_teachers'] + pisa['perform_poorly']
```

```python
#Create a new variable sense of belonging
agreement_dict = {'Strongly disagree':-2,'Disagree':-1,'Agree':1,'Strongly agree':2}

columnlist = [['feel_outsider','ST87Q01'],
              ['makes_friends','ST43Q02'],
              ['belong_at_school','ST87Q03'],
              ['awkward_at_school','ST87Q04'],
              ['liked_by_others','ST87Q05'],
              ['feel_lonely','ST87Q06'],
              ['feel_happy','ST87Q07'],
              ['ideal_at_school','ST87Q08'],
              ['satisfied_at_school','ST87Q09']]
                       
for newcolumn,mappedcolumn in columnlist:
    pisa[newcolumn] = pisa[mappedcolumn].map(agreement_dict)

pisa['sense_of_belonging'] = - pisa['feel_outsider'] + pisa['makes_friends'] + pisa['belong_at_school'] - pisa['awkward_at_school'] + pisa['liked_by_others'] - pisa['feel_lonely'] + pisa['feel_happy'] + pisa['ideal_at_school'] + pisa['satisfied_at_school']
```
